# Insurance Dataset - Exploratory Data Analysis

```python
#Importing libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Importing Datasets
data = pd.read_csv("Desktop/insurance.csv")
data.head()
```

# EDA

```python
# 1) let's answer the question of what's the average charges of male and female based on region ?

data.groupby(["sex","region"])["charges"].mean().unstack().plot(kind="bar",figsize=(10,5))
```

```python
# 2) Now based whether they smoke or not ?

data.groupby(["sex","region","smoker"])["charges"].mean().unstack().plot(kind="bar",figsize=(10,5))
```

```python
# 3) Let's find what's the average age of female and male based on region?

data.groupby(["sex","region"])["age"].mean().unstack().plot(kind="bar",figsize=(10,5))
```

```python
# 4) let's find out the mininum, maximum and mean BMI per region of female and male.

data.groupby(["sex","region"])["bmi"].agg(["min","max","mean"]).plot(kind="bar",figsize=(10,5))
```

```python
# 5) let's find out min and max age of person who smokes per region.

data[data["smoker"] == "yes"].groupby(["sex","region"])["age"].agg(["min","max"]).plot(kind="bar",figsize=(10,5))
```

```python
# 6) let's find out the total number of childer's person has based on region.

data.groupby(["sex","region"])["children"].count().unstack().plot(kind="bar",figsize=(10,5))
```

```python
# 7) Average Bmi per person based on region

data.groupby(["sex","region"])["bmi"].mean().unstack().plot(kind="bar",figsize=(10,5))
```

```python
# 8) Minimum, Maximum and Average charges of a person based on sex and smoking history.

data.groupby(["sex","smoker"])["charges"].agg(["min","max","mean"])
```
